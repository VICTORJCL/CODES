# Importando arquivos Excel

## Lendo arquivos xlsx

```python
import pandas as pd

tabela_vendas = pd.read_excel('datasets/vendas.xlsx')
tabela_vendas
```

```python
import pandas as pd

tabela_vendas = pd.read_excel('datasets/vendas.xlsx',
								index_col=0,
								decimal=',',
								usecols=[0, 1, 2, 3],
								)
print(tabela_vendas)
```

## Lendo arquivos csv

```python
import pandas as pd

tabela_vendas = pd.read_csv('datasets/vendas.csv')
tabela_vendas
```

```python
import pandas as pd

tabela_vendas = pd.read_csv('datasets/vendas.csv',
								index_col=0,
								sep=';',
								decimal=','
								)
tabela_vendas
```
